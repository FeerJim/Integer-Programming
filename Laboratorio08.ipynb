## Calendarización de Trabajos

**Breve intro al problema:** El problema de calendarización de trabajos busca optimizar la programación de tareas en un conjunto de máquinas, teniendo en cuenta restricciones relacionadas a la existencia de *tiempos de duración* para las tareas y de *tiempos de disponibilidad* para las máquinas.

**Problema:** Consideremos un aeropuerto con una única pista de aterrizaje y despegue, y un conjunto $A$ de aviones que requieren operar (aterrizar o despegar) en este aeropuerto. Cada avión $i \in A$ dispone de una **ventana de tiempo** $[a_i, b_i]$ dentro de la cual su operación de aterrizaje/despegue debe llevarse a cabo. Además, $t_i \in [a_i, b_i]$ representa el **instante de tiempo programado** (o deseado) para la operación del avión $i$.

El objetivo del problema es **minimizar los costos asociados a adelantos y retrasos** en las operaciones, cumpliendo los siguientes requerimientos:

- Cada avión $i \in A$ debe aterrizar o despegar dentro de su ventana de tiempo $[a_i, b_i]$.
- Si la operación del avión $i$ ocurre **después** del tiempo programado $t_i$, se incurre en un costo igual a $c_i$ por cada unidad de tiempo de retraso.
- Si la operación del avión $i$ ocurre **antes** de $t_i$, se incurre en un costo igual a $f_i$ por cada unidad de tiempo de adelanto.
- Por normas de seguridad aeronáutica, entre dos operaciones consecutivas de aterrizaje/despegue, debe existir una separación de al menos $s$ unidades de tiempo.

Este problema puede formularse como un problema de calendarización de trabajos, donde los trabajos representan las operaciones de los aviones y la máquina representa a la pista de aterrizaje.
